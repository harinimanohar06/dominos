```python
import pandas as pd
import numpy as np
pizza=pd.read_excel("C:/Users/HARINI/Downloads/Pizza_Sale.xlsx")
```

```python
pizza.isnull().sum()
```

```python
#null value column are in categorical column
for col in pizza.select_dtypes(include='object').columns:
    mode_val = pizza[col].mode(dropna=True)[0]
    pizza[col].fillna(mode_val, inplace=True)
```

```python
pizza.isna().sum()
```

```python
pizza['total_price']=pizza['total_price'].fillna(pizza['total_price'].median())
```

```python
# Convert price columns to float
pizza['unit_price'] = pd.to_numeric(pizza['unit_price'].astype(int),errors='coerce')
pizza['total_price'] = pd.to_numeric(pizza['total_price'].astype(int), errors='coerce')
```

```python

# Convert to datetime if not already
pizza['order_date'] = pd.to_datetime(pizza['order_date'], errors='coerce')

# Extract only the date part
pizza['order_date'] = pizza['order_date'].dt.date
```

```python
pizza["order_date"] = pd.to_datetime(pizza["order_date"], errors="coerce")
pizza["dayofweek"]=pizza["order_date"].dt.day_of_week
pizza['dayofyear']=pizza['order_date'].dt.day_of_year
pizza['dayofmonth']=pizza['order_date'].dt.days_in_month
pizza['month']=pizza['order_date'].dt.month
pizza['year']=pizza['order_date'].dt.year
pizza['quarter']=pizza['order_date'].dt.quarter
pizza['hour']=pizza['order_date'].dt.hour
```

```python
pizza
```

```python
df = pd.DataFrame(pizza, columns=['quantity','dayofweek'])
df

```

```python
df['rolling_sum'] = df['quantity'].rolling(window=4).sum()
df['rolling_mean'] = df['rolling_sum'].rolling(window=4).mean()
df
```

```python
df1=pd.DataFrame(df,columns=['rolling_sum','rolling_mean'])
df1
```

```python
df_concat=pd.concat([pizza,df1],axis=1)
df_concat
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = df_concat.select_dtypes(include=['object']).columns

# Create a new LabelEncoder instance
le_pizza = LabelEncoder()

# Apply LabelEncoder to each categorical column
label_encoders = {}
for column in categorical_cols:
    le = LabelEncoder()  # create a fresh encoder for each column
    df_concat[column] = le.fit_transform(df_concat[column])
    label_encoders[column] = le
df_concat.head()  
```

```python
df_concat.to_csv('pizza.csv', index=False)
```

```python
daily_sales = df_concat.groupby('order_date')['quantity'].sum()
```

visulation

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(daily_sales.index, daily_sales.values, color='blue', linewidth=2)

plt.title('Daily Pizza Sales', fontsize=20)
plt.xlabel('order_date')
plt.ylabel('Quantity Sold')
plt.grid(True)
plt.show()
```

```python
rolling_mean = daily_sales.rolling(window=30).mean()

plt.figure(figsize=(12,6))
plt.plot(daily_sales.index, daily_sales.values, label='Daily Sales', alpha=0.6)
plt.plot(rolling_mean.index, rolling_mean.values, color='red', label='30-day Rolling Mean')

plt.title('Daily Pizza Sales with Trend', fontsize=20)
plt.xlabel('order_date')
plt.ylabel('Quantity Sold')
plt.legend()
plt.show()
```

```python
category_sales = df_concat.groupby(['order_date','pizza_category'])['quantity'].count().unstack()

plt.figure(figsize=(12,6))
category_sales.plot(figsize=(12,6))
plt.title('Daily Sales by Pizza Category', fontsize=20)
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.legend(title='Category')
plt.show()
```

```python
# Rolling statistics (30-day window)
rolling_mean = daily_sales.rolling(window=30).mean()
rolling_std = daily_sales.rolling(window=30).std()

plt.figure(figsize=(12,6))
plt.plot(daily_sales, color='blue', label='Original')
plt.plot(rolling_mean, color='red', label='Rolling Mean (30 days)')
plt.plot(rolling_std, color='black', label='Rolling Std (30 days)')

plt.title('Stationarity Check: Rolling Mean & Std')
plt.xlabel('Date')
plt.ylabel('Quantity Sold')
plt.legend()
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(daily_sales.dropna())
print('ADF Statistic:', result[0])
print('p-value:', result[1])
```
